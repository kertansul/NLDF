```python
import cv2
import numpy as np
import sys
sys.path.insert(0,'../')
import NLDF
import os
import tensorflow as tf
import time
import vgg16
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
"""
ERROR will occur if this cell is run twice
"""

model = NLDF.Model()
model.build_model()

sess = tf.Session()
sess.run(tf.global_variables_initializer())
img_size = NLDF.img_size
label_size = NLDF.label_size

ckpt = tf.train.get_checkpoint_state('../Model/')
saver = tf.train.Saver()
saver.restore(sess, ckpt.model_checkpoint_path)

print 'Model Load Complete'
```

```python
filename='../../DeepSaliency/images/2891646_0.png'

img = cv2.imread( filename )

if img is not None:
    ori_img = img.copy()
    img_shape = img.shape
    img = cv2.resize(img, (img_size, img_size)) - vgg16.VGG_MEAN
    img = img.reshape((1, img_size, img_size, 3))

    start_time = time.time()
    result = sess.run(model.Prob,
                      feed_dict={model.input_holder: img})
    print("--- %s seconds ---" % (time.time() - start_time))

    result = np.reshape(result, (label_size, label_size, 2))
    result = result[:, :, 0]

    result = cv2.resize(np.squeeze(result), (img_shape[1], img_shape[0]))
    
    plt.subplot(1,2,1);plt.imshow(ori_img)
    plt.subplot(1,2,2);plt.imshow(result,cmap='gray')
```
